```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm

```

```python
dftrain=pd.read_csv('C:/Users/student/Downloads/AAATrain.csv')
```

```python
dftest=pd.read_csv('C:/Users/student/Downloads/AAATest.csv')
```

```python

model = sm.logit(formula='travel_destination_EUR ~  travel_travel_type_Ship + travel_revenue_index+\
travel_tripvalue_index + ers_services_count + one_booking+ travel_travel_month_8 +travel_travel_type_Insurance +\
travel_provide_name_best_western_inl + travel_channel_BRANCH + travel_channel_EVENT +\
travel_channel_WEB +travel_dom_int_D', data=dftrain).fit()
print(model.summary())
```

```python

Xtest = dftest[['travel_travel_type_Ship' , 'travel_revenue_index', 'travel_tripvalue_index',\
'ers_services_count','one_booking','travel_travel_month_8','travel_travel_type_Insurance',\
'travel_provide_name_best_western_inl','travel_channel_BRANCH','travel_channel_EVENT',\
'travel_channel_WEB','travel_dom_int_D']]
ytest = dftest['travel_destination_EUR']
  
total_yhat = model.predict(Xtest)
```

```python
dftest['total_yhat'] = total_yhat
```

```python
print(total_yhat)
```

```python

dftest = dftest.sort_values(by = 'total_yhat', ascending = False)
```

```python
dftest['Total_Quintile'] = pd.qcut(dftest['total_yhat'].rank(method='first'), 5, labels=[i for i in range (5, 0, -1)])
```

```python
print(dftest['Total_Quintile'])
```

```python
dftest['Total_Quintile'] = dftest['Total_Quintile'].astype('int')
```

```python
dftest['Total_Quintile'].value_counts().reset_index(name='count')
```

```python

europe_travelers = pd.crosstab(dftest['Total_Quintile'], dftest['travel_destination_EUR'], normalize='index')

```

```python
print(europe_travelers)
```
