```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("AAA_Travel_file2.csv")
```

```python
pd.set_option('display.max_rows', None)
```

```python
numerical_df = df.select_dtypes(include='number')

# Alternatively, you can combine both into a single DataFrame for better readability
min_max_df = pd.DataFrame({'min': numerical_df.min(), 'max': numerical_df.max()})
print("Minimum and maximum values of each numerical column:")
print(min_max_df)
```

```python
df['days_between'] = pd.cut(df['travel_daysbetween_booking_depart'],
                bins=[-float('Inf'), 31, 100, 250, float('Inf')],
                labels=['0-31 Days', '31-100 Days', '100-250 Days', '>250 Days'])
just_dummies = pd.get_dummies(df['days_between'], dtype=int)
df = pd.concat([df, just_dummies], axis=1) 
df['days_between'].describe()
```

```python
df[['0-31 Days', '31-100 Days', '100-250 Days', '>250 Days']].describe()
```

```python
df['rev_index'] = pd.cut(df['travel_revenue_index'],
                bins=[-1, 0, 12, 24, 36, float('Inf')],
                labels=['Missing_Trav_Rev', 'Quartile_1', 'Quartile_2', 'Quartile_3', 'Quartile_4'])
just_dummies = pd.get_dummies(df['rev_index'], dtype=int)
df = pd.concat([df, just_dummies], axis=1) 
df['rev_index'].describe()
```

```python
df[['Missing_Trav_Rev','Quartile_1', 'Quartile_2', 'Quartile_3', 'Quartile_4']].describe()
```

```python
np.random.seed(42)
dftrain = df.sample(frac=0.5,random_state=0)
dftest = df.drop(dftrain.index)
```

```python
dftrain.to_csv('AAATrain.csv')
dftest.to_csv('AAATest.csv')
```
