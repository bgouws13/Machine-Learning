```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
```

```python
dftrain=pd.read_csv('AAATrain.csv')
```

```python
dftest=pd.read_csv('AAATest.csv')
```

```python
model = sm.logit(formula='Florida ~ travel_total_bookings + travel_num_cancelled_bookings \
        + travel_daysbetween_booking_depart + travel_revenue_index + travel_tripvalue_index + \
        individual_tenure + individual_maritalstatus + individual_education + individual_children_7_9 \
        + days_between + Quartile_1 + Quartile_3', data=dftrain).fit()
print(model.summary())
```

```python
Xtest = dftest[['travel_total_bookings', 'travel_num_cancelled_bookings', 'travel_daysbetween_booking_depart', \
                'travel_revenue_index', 'travel_tripvalue_index', 'individual_tenure', \
                'individual_maritalstatus', 'individual_education', 'individual_children_7_9', \
                'days_between', 'Quartile_1', 'Quartile_3']]
ytest = dftest['Florida']
```

```python
Florida_yhat = model.predict(Xtest)
```

```python
dftest['Florida_yhat'] = Florida_yhat
```

```python
print(Florida_yhat)
```

```python
dftest = dftest.sort_values(by = 'Florida_yhat', ascending = False)
```

```python
dftest['Florida_Quintile'] = pd.qcut(dftest['Florida_yhat'], 5, labels=[i for i in range (5, 0, -1)])
```

```python
print(dftest['Florida_Quintile'])
```

```python
dftest['Florida_Quintile'] = dftest['Florida_Quintile'].astype('int')
```

```python
Florida_travelers = pd.crosstab(dftest['Florida_Quintile'], dftest['Florida'], normalize='index')
```

```python
print(Florida_travelers)
```

```python
dftest.to_csv('FLORIDAfile.csv')
```
