```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
dftrain=pd.read_csv('AAATrain.csv')
```

```python
dftest=pd.read_csv('AAATest.csv')
```

```python
model = smf.logit(formula='Florida ~ days_between + rev_index + individual_estimatedhouseholdincome + individual_tenure + individual_maritalstatus + individual_tenure + individual_education + individual_occupation', data=dftrain).fit()
print(model.summary())
```

```python
Xtest = dftest[['days_between', 'rev_index', 'individual_estimatedhouseholdincome', 'individual_education']]
ytest = dftest['Florida']
```

```python
dest1_yhat = model.predict(Xtest)
```

```python
dftest['dest1_yhat'] = dest1_yhat
```
