```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
dftrain=pd.read_csv('AAATrain.csv')
```

```python
dftest=pd.read_csv('AAATest.csv')
```

```python
columns_to_extract = [
    'days_between',
    'rev_index',
    'individual_estimatedhouseholdincome',
    'individual_maritalstatus',
    'individual_tenure',
    'individual_education',
    'individual_occupation'
]
```

```python
Xtest = dftest[columns_to_extract]

# Assuming 'Florida' is the target column
ytest = dftest['Florida']

# Print a success message
print("Columns extracted successfully.")
```

```python
model = smf.logit(formula='Florida ~ days_between + rev_index + individual_estimatedhouseholdincome + individual_tenure + individual_maritalstatus + individual_tenure + individual_education + individual_occupation + rev_index', data=dftrain).fit()
print(model.summary())
```

```python
Xtest = dftest[['days_between', + 'rev_index', 'individual_estimatedhouseholdincome', 'individual_maritalstatus', 'individual_tenure','individual_education', 'individual_occupation']]
ytest = dftest['Florida']
```

```python
dest1_yhat = model.predict(Xtest)
```

```python
dftest['dest1_yhat'] = dest1_yhat
```

```python
print(dest1_yhat)
```

```python

dftest = dftest.sort_values(by = 'dest1_yhat', ascending = False)
```

```python
dftest['dest1_Quintile'] = pd.qcut(dftest['dest1_yhat'], 5, labels=[i for i in range (5, 0, -1)])
```

```python
print(dftest['dest1_Quintile'])
```

```python
dftest['dest1_Quintile'] = dftest['dest1_Quintile'].astype('int')
```

```python
dest1_travelers = pd.crosstab(dftest['dest1_Quintile'], dftest['Florida'], normalize='index')
```

```python
print(dest1_travelers)
```
