```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
```

```python
dftrain=pd.read_csv('AAATrain.csv')
```

```python
dftest=pd.read_csv('AAATest.csv')
```

```python
model = sm.logit(formula='Florida ~ travel_total_bookings + travel_num_cancelled_bookings \
        + travel_daysbetween_booking_depart + travel_revenue_index + travel_tripvalue_index + individual_age + \
        individual_tenure + individual_maritalstatus + individual_education + individual_race + individual_children_7_9 \
        + days_between + Quartile_1 + Quartile_3', data=dftrain).fit()
print(model.summary())
```

```python
Xtest = dftest[['travel_total_bookings', 'travel_num_cancelled_bookings', 'travel_daysbetween_booking_depart', \
                'travel_revenue_index', 'travel_tripvalue_index', 'individual_age', 'individual_tenure', \
                'individual_maritalstatus', 'individual_education', 'individual_race', 'individual_children_7_9', \
                'days_between', 'Quartile_1', 'Quartile_3']]
ytest = dftest['Florida']
```

```python
dest1_yhat = model.predict(Xtest)
```

```python
dftest['dest1_yhat'] = dest1_yhat
```

```python
print(dest1_yhat)
```

```python
dftest = dftest.sort_values(by = 'dest1_yhat', ascending = False)
```

```python
dftest['dest1_Quintile'] = pd.qcut(dftest['dest1_yhat'], 5, labels=[i for i in range (5, 0, -1)])
```

```python
print(dftest['dest1_Quintile'])
```

```python
dftest['dest1_Quintile'] = dftest['dest1_Quintile'].astype('int')
```

```python
dest1_travelers = pd.crosstab(dftest['dest1_Quintile'], dftest['Florida'], normalize='index')
```

```python
print(dest1_travelers)
```
