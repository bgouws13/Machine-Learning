```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("AAA_Travel_file2.csv")
```

```python
df.describe(include='all')
```

```python
df['total_trips'] = pd.cut(df['travel_total_bookings'],
                      bins=[-2, 0, 2, float('Inf')],
                      labels=['bookings_missing', 'one_booking', 'more_than_one_booking'])
just_dummies = pd.get_dummies(df['total_trips'], dtype=int)
df = pd.concat([df, just_dummies], axis=1) 
df['total_trips'].describe()
```

```python
df[['bookings_missing', 'one_booking','more_than_one_booking']].describe()
```

```python
np.random.seed(42)
dftrain = df.sample(frac=0.5,random_state=0)
dftest = df.drop(dftrain.index)
```
