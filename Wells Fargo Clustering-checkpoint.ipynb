```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('WFTestNew.csv', low_memory=False)
```

```python
X = df[['total_yhat', 'muni_yhat', 'other_yhat']]
```

```python
# Initialize an empty list to store the within-cluster sum of squares (inertia) for each number of clusters
wcss = []
```

```python
# Determine the optimal number of clusters using the elbow method
for i in range(1, 11):  # Test cluster numbers from 1 to 10
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the within-cluster sum of squares (WCSS) vs. number of clusters
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

```python
# Establish 3 clusters using KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_
```
